## Class Methods

```python
class Person:
    def __init__(self, fname):
        self.name = fname

    def say_hi(self):
        print('Hello, my name is', self.name)
```

```python
p = Person('Anup')
print(p.name)
```

```python
p.say_hi()
```

```python
class Time:
    def __init__(self):
        self.hours =0
        self.minutes = 0

    def print_time(self):
        print( 'Hours:', self.hours, end=' ')
        print( 'Minutes:', self.minutes )
```

```python
time1 = Time()
time1.print_time()
time1.hours = 11
time1.minutes = 59
time1.print_time()
```

## Default values for attributes

```python
class Employee:
    def __init__(self, name, wage=8.25, hours=20):
        # Default employee is part time (20 hours/week)
        #and earns minimum wage

        self.name = name
        self.wage = wage
        self.hours = hours

todd = Employee('Todd') # Typical part-time employee
jason = Employee('Jason') # Typical part-time employee
tricia = Employee('Tricia', wage=12.50, hours=40)

# Manager employee
```

```python
employees = [todd, jason, tricia]

for e in employees:
    print(f'{e.name} earns {e.wage * e.hours :.2f} per week')
```
