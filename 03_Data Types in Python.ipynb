## Datatype : List

```python
# Some of the most expensive cars in the world
lamborghini_veneno = 3900000 # $3.9 million!
bugatti_veyron = 2400000     # $2.4 million!
aston_martin_one77 = 1850000 # $1.85 million!

prices = [lamborghini_veneno, bugatti_veyron, aston_martin_one77]

print('Lamborghini Veneno:', prices[0],'dollars')
print('Bugatti Veyron Super Sport:', prices[1], 'dollars')
print('Aston Martin One-77:',prices[2], 'dollars')

```

```python
# Demonstrate negative indexing on the prices list
print('Prices list:', prices)
print()

# Negative indexing from the end
print('prices[-1] (last element):', prices[-1])
print('prices[-2] (second last element):', prices[-2])
print('prices[-3] (third last element):', prices[-3])
print()

# Comparison with positive indexing
print('Positive vs Negative indexing:')
print(f'prices[0] = {prices[0]} is the same as prices[-3] = {prices[-3]}')
print(f'prices[1] = {prices[1]} is the same as prices[-2] = {prices[-2]}')
print(f'prices[2] = {prices[2]} is the same as prices[-1] = {prices[-1]}')
```

```python
my_list = [10, 'bw']
print(my_list)
my_list. append( 'abc' )
print( 'After append: ' , my_list)

my_list.pop(1)
print( 'After pop: ' , my_list)

my_list.remove('abc')
print( 'After remove:', my_list)
```

```python
# Concatenating lists
house_prices = [380000, 900000, 8875000] + [225000]
print('House prices:', house_prices)
print( 'There are', len(house_prices), 'prices in the list. ')

# Finding min, max
print( 'Cheapest house: ' ,min(house_prices))
print( 'Most expensive house: ' ,max(house_prices))
```

## Set

```python
# Create a set using the set() function.
nums1 = set([1,2,3])

# Create a setusing a set Literal.
nums2 = { 7,8,9 }

# Print the contents of the sets.
print(nums1)
print(nums2)
```

```python
# create sets
names1 = {'Pedro', 'Khan', 'Dean'}
names2 = {'Emilia','Kara','Tia'}

# Add element to set
names1.add('Hyungu')

print(names1)
```

```python
# Add names 2 to names 1
names1.update(names2)
print(names1)
```

```python
# Remove element from set
names1.remove('Dean')
print(names1)
```

```python
# Clear all elements from set
names2.clear()
print(names2)
```

```python
# Create sets
namesl ={'Corrin', 'Pedro', 'Khan', 'Dean'}
names2 ={'Emilia', 'Kara', 'Corrin','Tia'}
names3 ={'Karat', 'Kara', 'Karah', 'Khan'}
names4 ={'Khan', 'Dean', 'Pascale'}

# union namesl and names2
result_set = names1.union(names2)

print(f"Union of names1 and names2: {result_set}")
```

```python
# Intersection btwn result set and names3
result_set = result_set.intersection(names3)
print( f"Intersection btwn result_set and names3: {result_set}" )
```

```python
# Difference btwn result _ set and names4
result_set = result_set.difference(names4)
print( f"Difference btwn result _ set and names4: {result_set}" )
```
